# library import

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings(action='ignore')
import torch
import torch.nn as nn
import torch.nn.functional as F
import torchvision
import torchvision.transforms as transforms
from torch.utils.data import Dataset,DataLoader
import torch.optim as optim
import datatable as dt
import matplotlib.pyplot as plt
torch.manual_seed(1015)
# define 'device' to upload tensor in gpu
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
device
```

# 데이터 불러오기

```python
train_x = dt.open("../data/dacon/bit_trader/train_x_df.csv").to_pandas()
train_x = train_x[train_x['sample_id'] == 0]
train_x.head(2)
```

```python
train_y = dt.open("../data/dacon/bit_trader/train_y_df.csv").to_pandas()
train_y = train_y[train_y['sample_id'] == 0]
train_y.head(2)
```

# 데이터 수정

```python
train_x = train_x.drop(['sample_id', 'time', 'coin_index'], axis=1)
train_y = train_y.drop(['sample_id', 'time', 'coin_index'], axis=1)
train_x.head(2)
```

```python
# scaling
mini_x = train_x.iloc[:,:].min()
size_x = train_x.iloc[:,:].max() - train_x.iloc[:,:].min()
train_x.iloc[:,:] = (train_x.iloc[:,:] -  mini_x) / size_x
train_x.head(2)
```

```python
input_window = 30
output_window = 6

window_x = np.zeros((train_x.shape[0] - (input_window + output_window), input_window, len(train_x.columns)))
window_y = np.zeros((train_x.shape[0] - (input_window + output_window), output_window, len(train_x.columns)))

for start in range(train_x.shape[0] - (input_window + output_window)):
    end = start + input_window    
    window_x[start,:, :] = train_x.iloc[start : end                , : ].values
    window_y[start,:, :] = train_x.iloc[end   : end + output_window, : ].values
```

```python
window_x.shape, window_y.shape
```

# model 생성

```python
class LSTM(nn.Module):
    def __init__(self, input_size, hidden_size):
        super(LSTM, self).__init__()
        self.hidden_size = hidden_size
        self.lstm = nn.LSTM(input_size = input_size,
                            hidden_size = hidden_size,
                            batch_first=True)
        self.hidden_lstm = nn.LSTM(input_size = hidden_size,
                                   hidden_size = hidden_size,
                                   batch_first=True,
                                   bidirectional=False,
                                   num_layers=2)
        
        self.time_fc = nn.Linear(hidden_size, len(train_x.columns))
    
    def forward(self, x_time):
    
        out_time, _ = self.lstm(x_time)
        out_time, _ = self.hidden_lstm(out_time)
        out_time, _ = self.hidden_lstm(out_time)
#         out_time, _ = self.hidden_lstm(out_time)
#         out_time, _ = self.hidden_lstm(out_time)
        
        
        out_time = self.time_fc(out_time[:,-output_window:, :])
        
        return out_time.view(-1,output_window,len(train_x.columns))
    
model = LSTM(input_size = len(train_x.columns), hidden_size = 30).to(device)
```

# 학습

```python
device
```

```python
window_x = torch.tensor(window_x).float().to(device)
window_y = torch.tensor(window_y).float().to(device)

# Train model
optimizer = torch.optim.Adam(model.parameters(), lr = 1e-2)
criterion = nn.MSELoss(size_average = True)
num_epochs  = 100
```

```python
train_error = []
for t in range(num_epochs):
    train_pred = model(window_x)
    loss = criterion(train_pred, window_y) ### trend
    train_error.append(loss)
    
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    if t % 10 == 0 and t !=0:
        print(f"{t} Epochs train MSE: {loss.item():1.5f}")
```
